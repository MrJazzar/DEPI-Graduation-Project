# 2.  PREPROCESSING


```python
def preprocess(df: pd.DataFrame, target_col: str = None):
    """
    • Infer target column (last column if not given).
    • Impute: median for numeric, mode for categorical.
    • One-hot encode categorical features.
    • Label-encode target.
    Returns X, y_enc, le, numeric_cols, categorical_cols.
    """
    df = df.copy()

    if target_col is None:
        target_col = df.columns[-1]

    X = df.drop(columns=[target_col])
    y = df[target_col]

    numeric_cols     = X.select_dtypes(include=[np.number]).columns.tolist()
    categorical_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()

    for col in numeric_cols:
        X[col].fillna(X[col].median(), inplace=True)
    for col in categorical_cols:
        X[col].fillna(X[col].mode()[0], inplace=True)

    if categorical_cols:
        X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

    le    = LabelEncoder()
    y_enc = le.fit_transform(y)

    print(f"  Target column  : '{target_col}'")
    print(f"  Classes        : {list(le.classes_)}  →  {list(range(len(le.classes_)))}")
    print(f"  Feature count  : {X.shape[1]}")
    return X, y_enc, le, numeric_cols, categorical_cols
```

# 3.  FEATURE ENGINEERING


```python
def engineer_features(X: pd.DataFrame) -> pd.DataFrame:
    """
    Domain-inspired interaction features combining ideas from both scripts:
      • abs_yaw / abs_pitch   – magnitudes of head rotation
      • head_movement         – Euclidean (yaw² + pitch² + roll²)^0.5
      • low_ear               – binary: EAR < 0.25 (drowsiness cue)
      • high_gaze             – binary: |gaze| > 0.5 (strong deviation)
      • gaze_ear_ratio        – gaze × EAR interaction
      • distraction_score     – head_movement + |gaze| (composite index)
    """
    X = X.copy()

    if "yaw"   in X.columns: X["abs_yaw"]   = X["yaw"].abs()
    if "pitch" in X.columns: X["abs_pitch"] = X["pitch"].abs()

    if all(c in X.columns for c in ["yaw", "pitch", "roll"]):
        X["head_movement"] = np.sqrt(X["yaw"]**2 + X["pitch"]**2 + X["roll"]**2)

    if "ear"  in X.columns:
        X["low_ear"]  = (X["ear"] < 0.25).astype(int)

    if "gaze" in X.columns:
        X["high_gaze"] = (X["gaze"].abs() > 0.5).astype(int)

    if "gaze" in X.columns and "ear" in X.columns:
        X["gaze_ear_ratio"] = X["gaze"] * X["ear"]

    if "head_movement" in X.columns and "gaze" in X.columns:
        X["distraction_score"] = X["head_movement"] + X["gaze"].abs()

    print(f"  Features after engineering ({X.shape[1]}):")
    print(f"  {list(X.columns)}")
    return X
```

# 4.  CLASS BALANCING  (upsample minority)


```python
import numpy as np

def balance_classes(X: pd.DataFrame, y: np.ndarray) -> tuple:
    """Upsample minority class to match majority size."""
    X_c = X.copy()
    X_c["__target__"] = y

    majority_val = pd.Series(y).value_counts().idxmax()
    minority_val = pd.Series(y).value_counts().idxmin()

    df_maj = X_c[X_c["__target__"] == majority_val]
    df_min = X_c[X_c["__target__"] == minority_val]

    df_min_up = resample(df_min, replace=True,
                         n_samples=len(df_maj), random_state=42)

    df_bal = (pd.concat([df_maj, df_min_up])
                .sample(frac=1, random_state=42)
                .reset_index(drop=True))

    y_bal = df_bal["__target__"].values
    X_bal = df_bal.drop(columns=["__target__"])
    print(f"  Balanced size: {len(X_bal):,}  (original: {len(X):,})")
    return X_bal, y_bal
```

# 5.  SelectKBest


```python
def select_best_features(X_train, y_train, feature_cols):
    """Find optimal k via 5-fold CV on Logistic Regression F1."""
    k_values = range(2, X_train.shape[1] + 1)
    scores   = []

    for k in k_values:
        pipe = Pipeline([
            ("scale",  StandardScaler()),
            ("select", SelectKBest(f_classif, k=k)),
            ("model",  LogisticRegression(max_iter=1000))
        ])
        result = cross_val_score(pipe, X_train, y_train, cv=5, scoring="f1")
        scores.append(result.mean())

    best_k = list(k_values)[np.argmax(scores)]
    print(f"  Best k = {best_k}  |  Best CV F1 = {max(scores):.4f}")

    selector = SelectKBest(f_classif, k=best_k)
    X_train_sel = selector.fit_transform(X_train, y_train)
    X_test_mask = selector  # keep for test transform

    chosen = np.array(feature_cols)[selector.get_support()]
    print(f"  Selected features: {list(chosen)}")
    return selector, chosen, X_train_sel, best_k
```

# 6.  CLASS WEIGHTS


```python
def get_class_weights(y_train):
    """Compute balanced class weights (from Code 1)."""
    weights = compute_class_weight("balanced",
                                   classes=np.unique(y_train),
                                   y=y_train)
    cw = dict(zip(np.unique(y_train), weights))
    for c, w in cw.items():
        print(f"  Class {c}: weight = {w:.3f}")
    return cw
```

# 7.  BROAD MODEL COMPARISON  (AUC)


```python
def broad_model_comparison(X_train_sel, X_test_sel, y_train, y_test, cw):
    """
    Train 8–9 classifiers with cross-validation and report
    CV F1, Accuracy, F1, AUC in a formatted table (Code 1 style).
    Returns dict of fitted models.
    """
    models = {
        "Logistic"    : LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42),
        "DecisionTree": DecisionTreeClassifier(class_weight="balanced", random_state=42),
        "RandomForest": RandomForestClassifier(n_estimators=100, class_weight="balanced", random_state=42),
        "GradientBoost": GradientBoostingClassifier(random_state=42),
        "AdaBoost"    : AdaBoostClassifier(random_state=42),
        "SVM"         : SVC(kernel="rbf", probability=True, class_weight="balanced", random_state=42),
        "KNN"         : KNeighborsClassifier(),
        "NaiveBayes"  : GaussianNB(),
    }
    if HAS_XGB:
        models["XGBoost"] = XGBClassifier(
            use_label_encoder=False, eval_metric="logloss",
            scale_pos_weight=cw.get(0, 1) / cw.get(1, 1),
            random_state=42
        )

    scale_needed = {"Logistic", "SVM", "KNN", "NaiveBayes"}
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    print("-" * 70)
    print(f"{'Model':<18}{'CV F1':>10}{'ACC':>10}{'F1':>10}{'AUC':>10}")
    print("-" * 70)

    fitted_models = {}
    results_list  = []

    for name, model in models.items():
        if name in scale_needed:
            sc  = StandardScaler()
            Xtr = sc.fit_transform(X_train_sel)
            Xte = sc.transform(X_test_sel)
        else:
            Xtr, Xte = X_train_sel, X_test_sel

        cv_f1 = cross_val_score(model, Xtr, y_train, cv=cv, scoring="f1").mean()
        model.fit(Xtr, y_train)
        pred  = model.predict(Xte)
        acc   = accuracy_score(y_test, pred)
        f1    = f1_score(y_test, pred, zero_division=0)
        auc   = roc_auc_score(y_test, model.predict_proba(Xte)[:, 1]) \
                if hasattr(model, "predict_proba") else 0.0

        print(f"{name:<18}{cv_f1:>10.4f}{acc:>10.4f}{f1:>10.4f}{auc:>10.4f}")

        fitted_models[name] = model
        results_list.append({"Model": name, "CV_F1": cv_f1,
                              "Accuracy": acc, "F1": f1, "AUC": auc})

    print("-" * 70)
    return fitted_models, pd.DataFrame(results_list)
```

# 8.  FINE-TUNED PIPELINES


```python
def build_pipelines(numeric_features: list) -> dict:
    """SVM + RandomForest with full ColumnTransformer pipelines."""
    preprocessor = ColumnTransformer(
        [("scaler", StandardScaler(), numeric_features)],
        remainder="passthrough"
    )
    return {
        "SVM": (
            Pipeline([("prep", preprocessor),
                      ("clf",  SVC(probability=True, random_state=42))]),
            {"clf__C": [0.1, 1, 10, 100],
             "clf__kernel": ["rbf", "linear"],
             "clf__gamma":  ["scale", "auto"]}
        ),
        "RandomForest": (
            Pipeline([("prep", preprocessor),
                      ("clf",  RandomForestClassifier(random_state=42))]),
            {"clf__n_estimators":     [100, 200, 300],
             "clf__max_depth":        [5, 10, 15, None],
             "clf__min_samples_split":[2, 5, 10],
             "clf__max_features":     ["sqrt", "log2"]}
        ),
    }


def tune_model(name, pipeline, param_grid, X_train, y_train, cv=5):
    """GridSearchCV tuning optimised for f1_macro."""
    cv_strategy = StratifiedKFold(n_splits=cv, shuffle=True, random_state=42)
    grid = GridSearchCV(pipeline, param_grid, cv=cv_strategy,
                        scoring="f1_macro", n_jobs=-1, refit=True)
    print(f"\n  Tuning {name} …")
    grid.fit(X_train, y_train)
    print(f"  ✓ Best params : {grid.best_params_}")
    print(f"  ✓ Best CV F1  : {grid.best_score_:.4f}")
    return grid


def evaluate_model(name, fitted_grid, X_test, y_test, le):
    """Compute Accuracy / Precision / Recall / F1-macro on test set."""
    y_pred = fitted_grid.predict(X_test)
    metrics = {
        "Model"    : name,
        "Accuracy" : accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, average="macro", zero_division=0),
        "Recall"   : recall_score(y_test, y_pred,    average="macro", zero_division=0),
        "F1-Score" : f1_score(y_test, y_pred,        average="macro", zero_division=0),
    }
    print(f"\n  ── {name} ──")
    for k, v in metrics.items():
        if k != "Model": print(f"  {k:<12}: {v:.4f}")
    print()
    print(classification_report(y_test, y_pred,
          target_names=le.classes_, digits=4))
    return metrics
```

# 10.  SHAP ANALYSIS

```python
def run_shap_analysis(rf_grid, X_test, output_dir, max_display=10):
    if not HAS_SHAP:
        print("  SHAP skipped – library not installed.")
        return
    print("  Running SHAP analysis …")
    rf_clf  = rf_grid.best_estimator_.named_steps["clf"]
    prep    = rf_grid.best_estimator_.named_steps["prep"]
    X_test_tr = prep.transform(X_test)
    feat_names = prep.get_feature_names_out()
    explainer   = shap.TreeExplainer(rf_clf)
    shap_values = explainer.shap_values(X_test_tr)
    # For binary, shap_values may be a list
    sv = shap_values[1] if isinstance(shap_values, list) else shap_values
    fig, ax = plt.subplots(figsize=(10, 6))
    shap.summary_plot(sv, X_test_tr, feature_names=feat_names,
                      max_display=max_display, show=False)
    path = os.path.join(output_dir, "09_shap_summary.png")
    plt.savefig(path, dpi=150, bbox_inches="tight"); plt.close()
    print(f"  SHAP summary saved → {path}")
```

# 11.  ABLATION STUDY

```python
def ablation_study(X_train, X_test, y_train, y_test,
                   imp_series, le, output_dir):
    print("\n  ── Ablation Study ──")
    ranked = [f.replace("scaler__","").replace("remainder__","")
              for f in imp_series.index.tolist()]
    drop_order      = ranked[::-1]
    current_features = list(X_train.columns)
    records = []

    for drop_feat in drop_order:
        if drop_feat not in current_features: continue
        current_features = [f for f in current_features if f != drop_feat]
        if not current_features: break

        Xtr = X_train[current_features]
        Xte = X_test[current_features]
        num_sub = [c for c in current_features if X_train[c].dtype in
                   [np.float64, np.float32, np.int64]]

        prep_sub = ColumnTransformer(
            [("scaler", StandardScaler(), num_sub)], remainder="passthrough")
        pipe_sub = Pipeline([("prep", prep_sub),
                              ("clf", RandomForestClassifier(n_estimators=100, random_state=42))])
        pipe_sub.fit(Xtr, y_train)
        f1 = f1_score(y_test, pipe_sub.predict(Xte), average="macro")
        records.append({"Dropped": drop_feat,
                        "Remaining": len(current_features),
                        "F1_macro": round(f1, 4)})
        print(f"    Dropped '{drop_feat}' → {len(current_features)} feats | F1={f1:.4f}")

    df_abl = pd.DataFrame(records)
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(df_abl["Remaining"], df_abl["F1_macro"], marker="o",
            lw=2, color="darkorange")
    ax.set_xlabel("Remaining Features"); ax.set_ylabel("F1-Score (macro)")
    ax.set_title("Ablation Study – Feature Removal Impact", fontsize=13, fontweight="bold")
    ax.invert_xaxis(); ax.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    path = os.path.join(output_dir, "10_ablation_study.png")
    fig.savefig(path, dpi=150, bbox_inches="tight"); plt.close(fig)
    print(f"  Saved → {path}")
    return df_abl
```

# 12.  ENGAGEMENT SCORING 


```python
def build_engagement_scorer(best_pipeline, le,
                             high_thresh=0.70, low_thresh=0.40):
    """
    Returns score_engagement(X_new) callable.
    Tiers:  High   ≥ high_thresh
            Medium ≥ low_thresh
            Low    < low_thresh
    """
    focused_idx = list(le.classes_).index("focused")

    def score_engagement(X_new: pd.DataFrame) -> pd.DataFrame:
        proba    = best_pipeline.predict_proba(X_new)
        p_focused = proba[:, focused_idx]
        tiers = np.where(p_focused >= high_thresh, "High",
                np.where(p_focused >= low_thresh,  "Medium", "Low"))
        return pd.DataFrame({"P(focused)": p_focused.round(4),
                              "EngagementTier": tiers},
                             index=X_new.index)

    return score_engagement
```

```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder

def main(filepath:   str = "/mnt/user-data/uploads/features_clean.csv",
         output_dir = "outputs"):

    os.makedirs(output_dir, exist_ok=True)

    # ── STEP 1: Load & Explore ────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 1 : LOAD & EXPLORE DATA")
    print("═"*65)
    df_raw = load_and_explore(filepath)

    # ── STEP 2: Preprocess ────────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 2 : PREPROCESS")
    print("═"*65)
    X, y, le, numeric_cols, categorical_cols = preprocess(df_raw)

    # ── STEP 3: Feature Engineering ──────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 3 : FEATURE ENGINEERING")
    print("═"*65)
    X = engineer_features(X)
    all_numeric = X.select_dtypes(include=[np.number]).columns.tolist()

    # Attach encoded label for visualisation purposes
    df_proc = X.copy()
    df_proc["label_encoded"] = y

    # ── STEP 4: EDA Plots ─────────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 4 : EDA VISUALISATIONS")
    print("═"*65)
    plot_target_distribution(df_raw, output_dir)
    plot_feature_distributions(df_proc, output_dir)
    plot_boxplots(df_proc, output_dir)

    # ── STEP 5: Class Balancing ───────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 5 : CLASS BALANCING")
    print("═"*65)
    X_bal, y_bal = balance_classes(X, y)

    # ── STEP 6: Train / Test Split ────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 6 : TRAIN / TEST SPLIT  (80/20, stratified)")
    print("═"*65)
    X_train, X_test, y_train, y_test = train_test_split(
        X_bal, y_bal, test_size=0.2, random_state=42, stratify=y_bal
    )
    print(f"  Train: {X_train.shape}  |  Test: {X_test.shape}")

    # ── STEP 7: Class Weights ─────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 7 : CLASS WEIGHTS")
    print("═"*65)
    cw = get_class_weights(y_train)

    # ── STEP 8: SelectKBest ───────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 8 : SELECTKBEST FEATURE SELECTION")
    print("═"*65)
    feature_cols = list(X_train.columns)
    selector, chosen_features, X_train_sel, best_k = \
        select_best_features(X_train.values, y_train, feature_cols)
    X_test_sel = selector.transform(X_test.values)

    # ── STEP 9: Broad Model Comparison ───────────────────────────
    print("\n" + "═"*65)
    print("  STEP 9 : BROAD MODEL COMPARISON  (9 classifiers)")
    print("═"*65)
    fitted_models, df_broad = broad_model_comparison(
        X_train_sel, X_test_sel, y_train, y_test, cw)
    plot_broad_comparison(df_broad, output_dir)

    # ── STEP 10: Hyperparameter Tuning (SVM + RF pipelines) ──────
    print("\n" + "═"*65)
    print("  STEP 10: HYPERPARAMETER TUNING  (GridSearchCV, CV=5)")
    print("═"*65)
    models_dict  = build_pipelines(all_numeric)
    fitted_grids = {}
    for name, (pipe, param_grid) in models_dict.items():
        fitted_grids[name] = tune_model(
            name, pipe, param_grid, X_train, y_train, cv=5)

    # ── STEP 11: Evaluate Tuned Models ───────────────────────────
    print("\n" + "═"*65)
    print("  STEP 11: EVALUATE TUNED MODELS")
    print("═"*65)
    results = []
    for name, grid in fitted_grids.items():
        results.append(evaluate_model(name, grid, X_test, y_test, le))

    df_metrics = pd.DataFrame(results).set_index("Model")
    best_name  = df_metrics["F1-Score"].idxmax()
    print(f"\n   Best tuned model (F1-macro): {best_name}")
    print(df_metrics.to_string(float_format="{:.4f}".format))

    # ── STEP 12: Confusion Matrix ─────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 12: CONFUSION MATRIX")
    print("═"*65)
    plot_confusion_matrix(fitted_grids[best_name],
                          X_test, y_test, le, best_name, output_dir)

    # ── STEP 13: ROC Curves ───────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 13: ROC CURVES")
    print("═"*65)
    plot_roc_curves(fitted_grids, X_test, y_test, output_dir)

    # ── STEP 14: Feature Importance ──────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 14: FEATURE IMPORTANCE  (Random Forest)")
    print("═"*65)
    imp_series = plot_feature_importance(fitted_grids["RandomForest"], output_dir)

    # ── STEP 15: SHAP ─────────────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 15: SHAP ANALYSIS  (Random Forest)")
    print("═"*65)
    run_shap_analysis(fitted_grids["RandomForest"], X_test, output_dir)

    # ── STEP 16: Ablation Study ───────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 16: ABLATION STUDY")
    print("═"*65)
    df_abl = ablation_study(X_train, X_test, y_train, y_test,
                             imp_series, le, output_dir)

    # ── STEP 17: Dashboard ────────────────────────────────────────
    print("\n" + "═"*65)
    print("  STEP 17: MODEL COMPARISON DASHBOARD")
    print("═"*65)
    plot_dashboard(df_metrics, output_dir)

    # ── STEP 18: Engagement Scoring Demo ─────────────────────────
    print("\n" + "═"*65)
    print("  STEP 18: ENGAGEMENT SCORING DEMO")
    print("═"*65)
    score_engagement = build_engagement_scorer(
        fitted_grids[best_name], le, high_thresh=0.70, low_thresh=0.40)
    demo   = X_test.head(10)
    scored = score_engagement(demo)
    print("  Sample scoring on 10 test rows:")
    print(scored.to_string())
    print("\n  Tier thresholds:")
    print("    High   → P(focused) ≥ 0.70  – actively engaged")
    print("    Medium → 0.40 ≤ P(focused) < 0.70  – partially engaged")
    print("    Low    → P(focused) < 0.40  – distracted / disengaged")

    # ── Save outputs ──────────────────────────────────────────────
    df_metrics.to_csv(os.path.join(output_dir, "model_metrics_tuned.csv"))
    df_broad.to_csv(os.path.join(output_dir, "model_metrics_broad.csv"), index=False)
    df_abl.to_csv(os.path.join(output_dir, "ablation_results.csv"), index=False)
    scored.to_csv(os.path.join(output_dir, "engagement_scores_sample.csv"))

    print("\n" + "═"*65)
    print("    PIPELINE COMPLETE")
    print(f"  Best model   : {best_name}")
    print(f"  All outputs  : {output_dir}")
    print("═"*65)

    return {
        "best_model_name"  : best_name,
        "best_model_grid"  : fitted_grids[best_name],
        "score_engagement" : score_engagement,
        "metrics_tuned"    : df_metrics,
        "metrics_broad"    : df_broad,
        "ablation"         : df_abl,
        "fitted_grids"     : fitted_grids,
        "fitted_models"    : fitted_models,
    }


# ── Entry Point ───────────────────────────────────────────────────
if __name__ == "__main__":
    results = main()
```
